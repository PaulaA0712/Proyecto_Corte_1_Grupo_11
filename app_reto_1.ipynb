# Reto 1: LOGÍSTICA - Ruta Óptima

```python
"""
RUTA-ÓPTIMA

Este programa permite ingresar el peso y destino de 5 paquetes,
clasificarlos automáticamente según su peso, calcular su costo
de envío y generar un manifiesto ordenado por peso.
"""

# Inicio del sistema

paquetes = []  # Lista para almacenar los paquetes
tarifa_base = 20000
opcion = 0

# Menú Interactivo (while)

while opcion != 3:
    
    print("\n===== MENÚ RUTA ÓPTIMA =====")
    print("1. Ingresar paquete")
    print("2. Ver manifiesto de carga")
    print("3. Salir")
    
    try:
        opcion = int(input("Seleccione una opción: "))
    except:
        print("Entrada inválida.")
        continue
    
    
    # OPCIÓN 1: INGRESAR PAQUETE
    
    if opcion == 1:
        
        if len(paquetes) < 5:   # Validación de mínimo 5 registros
            
            try:
                peso = float(input("Ingrese el peso del paquete (kg): "))
                
                if peso <= 0:
                    print("Error: El peso debe ser mayor que 0.")
                    continue
                
                destino = input("Ingrese el destino del paquete: ")
                
                # Clasificación automática (if / elif / else)
                if peso < 1:
                    tipo = "Documento"
                elif peso < 10:
                    tipo = "Paquetería"
                else:
                    tipo = "Carga"
                
                # Cálculo del costo
                costo = tarifa_base * peso
                
                # Guardar datos en lista (diccionario)
                paquetes.append({
                    "peso": peso,
                    "destino": destino,
                    "tipo": tipo,
                    "costo": costo
                })
                
                print("Paquete registrado exitosamente.")
            
            except:
                print("Error en el ingreso de datos.")
        
        else:
            print("Ya se registraron los 5 paquetes permitidos.")
    

    # OPCIÓN 2: VER MANIFIESTO

    
    elif opcion == 2:
        
        if len(paquetes) < 5:
            print(f"Debe registrar los 5 paquetes. Faltan {5 - len(paquetes)}")
        
        else:
            paquetes.sort(key=lambda x: x["peso"])
            
            print("\n===== MANIFIESTO DE CARGA =====")
            
            total_envios = 0
            
            # Ciclo for obligatorio
            for paquete in paquetes:
                
                print(f"Destino: {paquete['destino']}")
                print(f"Peso: {paquete['peso']} kg")
                print(f"Tipo: {paquete['tipo']}")
                print(f"Costo: ${paquete['costo']:,.0f}")
                print("---------------------------")
                
                total_envios += paquete["costo"]
            
            print(f"TOTAL GENERAL: ${total_envios:,.0f}")
    
    
    # OPCIÓN 3: SALIR
    elif opcion == 3:
        if len(paquetes) < 5:
            print(f"No puedes salir. Faltan {5 - len(paquetes)} paquetes por registrar.")
            opcion = 0  # fuerza a que el while continúe
        else:
            print("Programa finalizado.")
```
